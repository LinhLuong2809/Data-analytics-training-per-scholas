```python
'''
Task 1: Inheritance - Building a Vehicle Hierarchy

In this task, learners will dive into the concept of Inheritance by building a hierarchy of vehicle classes.

Define a base class, `Vehicle`, with common attributes like `make`, `model`, and `year`.
Create subclasses such as `Car`, `Motorcycle`, and `Truck` that inherit from the `Vehicle` class.
Extend or modify functionality in the subclasses to represent unique characteristics of each vehicle type.
Instantiate objects of each subclass and showcase the use of inherited attributes and methods.

'''

class Vehicle():
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def sound(self):
        pass
        
class Car(Vehicle):
    def __init__(self, make , model, year, doors):
        super().__init__(make,model,year)
        self.doors = doors
        
    def sound(self):
        print("Broom broom!")

class Truck(Vehicle):
    def __init__(self, make , model, year, wheel):
        super().__init__(make,model,year)
        self.wheel = wheel
    
    def sound(self):
        print("BRRRROOOOOOOOOM!")
            
class Motorcycle(Vehicle):
    def __init__(self, make , model, year, paint):
        super().__init__(make,model, year)
        self.paint = paint
        
    def sound(self):
        print("Broooooo!")
        
toyota = Car("Toyota", "Campry", 2023, 4)
print(f"Car: {toyota.make} {toyota.model} {toyota.year} {toyota.doors} doors")
Chevrolet = Truck("Chevrolet", "Colorado", 2023, 17)
print(f"Truck: {Chevrolet.make} {Chevrolet.model} {Chevrolet.year} {Chevrolet.wheel} inches wheels")
honda = Motorcycle("Honda", "Navi", 2024, "Red")
print(f"Motorcycle: {honda.make} {honda.model} {honda.year} {honda.paint} body paint")

toyota.sound()
Chevrolet.sound()
honda.sound()
```

```python
'''
Task 3: Polymorphism - Implementing a Shape Hierarchy

In this task, learners will explore Polymorphism by implementing a hierarchy of geometric shapes.

Define a base class, `Shape`, with a method like `calculate_area`.
Create subclasses such as `Circle`, `Rectangle`, and `Triangle` that inherit from the `Shape` class.
Override the `calculate_area` method in each subclass to provide unique implementations.
Showcase polymorphic behavior by using a list of shapes and calculating their areas.

'''

class Shape:
    def __init__(self):
        pass
    
    def calculate_area(self):
        pass

class Circle:
    def __init__(self, radius):
        super().__init__()
        self.radius = radius
        
    def calculate_area(self):
        area = self.radius**2 * 3.14
        return area

class Rectangle:
    def __init__(self, length, width):
        super().__init__()
        self.length = length
        self.width = width
        
    def calculate_area(self):
        area = self.length * self.width
        return area
    
class Triangle:
    def __init__(self, base, heigh):
        super().__init__()
        self.base = base
        self.heigh = heigh
        
    def calculate_area(self):
        area = (self.base * self.heigh)/2
        return area
    
circle = Circle(5)
rect = Rectangle(3,9)
tri = Triangle(4,5)

print(f"Circle radius is {circle.radius}. Its area is {round(circle.calculate_area(),1)}")
print(f"Rectangle has length {rect.length} and width {rect.width}. Its area is {rect.calculate_area()}")
print(f"Triangle has base of {tri.base} and heigh of {tri.heigh}. Its area is {tri.calculate_area()}")
```
